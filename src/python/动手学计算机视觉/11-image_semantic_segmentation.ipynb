# 第十一章 语义分割  
## 11.1 概述  
- 语义分割是一种像素级的"图像分类"  
- 在语义分割中，通常使用平均交并比(mean intersection over union, mIoU)作为评测指标。  
- **全卷积网络(FCN)**与用于分类的CNN的区别在于，网络输出的全连接层FC被替换为卷积核大小为1*1的卷积层，由此，将卷积层输出的多通道特征图映射成一个空间分辨率不变的通道数为$|C|$的类别置信图，也叫做**热图(heatmap)**，例如
  $$
  h*w*c -> h*w*c_numclasses
  $$
  c_numclasses为像素类别个数。  
- 随着卷积的进行，特征图越来越小，最终输出的热图要比原图分辨率小的多，无法精准地勾勒出原图中物体的轮廓，为了解决以上问题，引入两种架构，**上采样(upsampling)**和**跳跃连接(skip-connection)**  
## 11.2 上采样  
- 使用双线性插值法，公式变形后，可由卷积操作进行实现  
- 用于特征图上采样的卷积成为**转置卷积**或**反卷积**，模拟了卷积的反向过程。方法为torch.nn.ConvTranspose2d  
- 也可用传统的插值方式，例如F.interpolation  

## 11.3 跳跃连接  
- 浅层卷积层的特征图较大，能够精确定位物体的边界，但是语义性不够；深层卷积层的特征图较小，但是语义性很强。  
- FCN通过跳跃连接将网络中不同层级的卷积层特征融合，使得得到较大的特征图的同时，也能使特征图拥有丰富的语义信息。
